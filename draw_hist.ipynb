```python
import ROOT
# Access the data file and ntuple inside of it  
data_file = ROOT.TFile.Open('data.root')
signal_tree= data_file.Get("signal")
background_tree= data_file.Get("background")
# Create a empty 1D histogram to draw jet pTs
h_signal_mass = ROOT.TH1F("h_signal_mass", ";m_{bb}; Entries", 100, 0, 200)
h_background_mass = ROOT.TH1F("h_background_mass", ";m_{bb}; Entries", 100, 0, 200)
# Fill the signal and background events
signal_tree.Draw("m_bb>>h_signal_mass", "", "gof")
background_tree.Draw("m_bb>>h_background_mass", "", "gof")
# create a new TCanvas
c = ROOT.TCanvas()
#Define marker style, color and size for signal, respectively. 
h_signal_mass.SetLineColor(ROOT.kRed)
h_signal_mass.SetStats(0) # Close info box of plot
h_signal_mass.Draw() # Draw signal distribution to canvas
#Define marker style, color and size for background, respectively. 
h_background_mass.SetLineColor(ROOT.kBlue)
h_background_mass.SetStats(0) # Close info box of plot
h_background_mass.Draw("same");  # Draw background distribution to canvas
# Legend box definition which helps us to know which is which
legend = ROOT.TLegend(0.65,0.77,0.86,0.89)
legend.AddEntry(h_signal_mass, "Signal", "l") #Introduce the signal histogram to legend as a point
legend.AddEntry(h_background_mass, "Background", "l") #Introduce the background histogram to legend as a point
legend.Draw("sames") #Draw to well defined legend to canvas
c.Draw()
```
